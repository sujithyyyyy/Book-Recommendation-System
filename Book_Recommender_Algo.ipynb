````python
def recommend(book_name):
  """
  Recommends books similar to a given book title.

  Args:
    book_name (str): The title of the book for which to generate recommendations.

  Returns:
    list: A list of lists, where each inner list contains information about a
          recommended book (title, author, and image URL).
  """
  # Find the index of the given book title in the pivot table (pt).
  index = np.where(pt.index == book_name)[0][0]

  # Get the similarity scores for the given book and sort them in descending order.
  # Select the top 5 most similar books (excluding the book itself).
  similar_items = sorted(list(enumerate(similarity_scores[index])),
                         key=lambda x: x[1],
                         reverse=True)[1:6]

  # Initialize an empty list to store the recommendations.
  data = []

  # Iterate through the top 5 similar books and extract their information.
  for i in similar_items:
    item = []  # Initialize an empty list to store information about each book.

    # Get the book information from the 'books' DataFrame based on the book title.
    temp_df = books[books['Book-Title'] == pt.index[i[0]]]

    # Add the book title, author, and image URL to the 'item' list.
    item.extend(list(temp_df.drop_duplicates('Book-Title')["Book-Title"].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')["Book-Author"].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')["Image-URL-M"].values))

    # Add the 'item' (book information) to the 'data' list.
    data.append(item)

  # Return the list of recommended books.
  return data
````
